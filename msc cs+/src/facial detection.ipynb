```python
import numpy as np
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr
import matplotlib.pyplot as plt
```

```python
def calculate_rmse(predictions, targets):
    mse = mean_squared_error(targets, predictions)
    rmse = np.sqrt(mse)
    return rmse
```

```python
def calculate_ccc(predictions, targets):
    mean_predictions = np.mean(predictions)
    mean_targets = np.mean(targets)

    cov_matrix = np.cov(predictions, targets)
    cov_xy = cov_matrix[0, 1]

    std_predictions = np.std(predictions)
    std_targets = np.std(targets)

    ccc = 2 * cov_xy / (std_predictions**2 + std_targets**2 + (mean_predictions - mean_targets)**2)
    return ccc
```

```python
np.random.seed(42)
predictions = np.random.rand(100)
targets = predictions + np.random.normal(0, 0.1, size=100)

# Calculate RMSE and CCC
rmse = calculate_rmse(predictions, targets)
ccc = calculate_ccc(predictions, targets)
```

```python
print("Root Mean Squared Error (RMSE):", rmse)
print("Concordance Correlation Coefficient (CCC):", ccc)
```

```python
plt.figure(figsize=(10, 5))
```
